```python
import os
import pandas as pd
import warnings
warnings.simplefilter("ignore")
#lst=["01_JAMMU & KASHMIR_APR'23","03_PUNJAB_MAY'23"]
df_blank=pd.DataFrame()
lst=os.listdir("C:/Users/Muskan.Kalra/Documents/GST files")
for items in lst:
    #print(items)
    try:
        df=pd.read_excel("C:/Users/Muskan.Kalra/Documents/GST files/"+items,"Trial")
        df.columns=["Company GSTIN","Company","Month","SRNo","Receiver GSTIN","Taxable Value","IGST","CGST","SGST"]
        df_temp=df_blank.append(df)
        df_blank=df_temp
    except:
        print(items)
```

```python
df_blank=pd.read_csv("C:/Users/Muskan.Kalra/Documents/GST files/GST_Blank.csv")
```

```python
df_blank=df_blank.reset_index(drop=True)
df_blank["Company GSTIN"] = df_blank["Company GSTIN"].str.strip()
#df_blank.to_csv("C:/Users/Muskan.Kalra/Documents/GST files/GST_Final.csv")
df_blank['Company State Initial'] = df_blank['Company GSTIN'].str[:2]
df_blank['Receiver State Initial'] = df_blank['Receiver GSTIN'].str[:2]
df_blank["Company Pan"]=df_blank['Company GSTIN'].str[2:12]
df_blank["Receiver Pan"]=df_blank['Receiver GSTIN'].str[2:12]
df_state_code=pd.read_excel("C:/Users/Muskan.Kalra/Documents/state code list GST.xlsx")
df_state_code=df_state_code[["State","State code list under GST"]]
df_state_code.rename(columns = {'State':'Company State Initial'}, inplace = True)
df_blank["Company State Initial"]=df_blank["Company State Initial"].astype(int)
df_final=pd.merge(df_blank,df_state_code,how="left",on="Company State Initial")
df_final=df_final.reset_index(drop=True)
df_final.loc[df_final["Company Pan"]==df_final["Receiver Pan"],"Remarks"]="STO"
```
