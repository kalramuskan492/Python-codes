```python
import pandas as pd
import datetime
import time
import re
agentBreaks=pd.read_excel("Not Ready Reason Codes By Agent.xlsx", engine='openpyxl')
```

```python
if (agentBreaks.shape[1] == 6):
    agentBreaks.columns = ['2','3','4','5','6','7']
else:
    agentBreaks.columns = ['1','2','3','4','5','6','7']
```

```python
row=0
agentName='Undefined'
dt='Undefined'
val=""

f=open('tmp.csv','w')
```

```python
f.write('Agent Name,Date,Type,Duration\n'); f.flush()


for index in agentBreaks.index:
    #print(index,agentBreaks.loc[index,'2'],agentBreaks.loc[index,'3'],agentBreaks.loc[index,'4'])
    if (index==0):
        if (agentBreaks.loc[index,'3'] !='Not Ready Reason Codes By Agent'):
            print("Invalid File : 'Not Ready Reason Codes By Agent.xlsx'")
            exit(1)
    if (not pd.isnull(agentBreaks.loc[index,'3']) and isinstance(agentBreaks.loc[index,'3'], str)):
        val=agentBreaks.loc[index,'3'][0:10]
        
    if (not pd.isnull(agentBreaks.loc[index,'3']) and isinstance(agentBreaks.loc[index,'3'], datetime.date) and pd.isnull(agentBreaks.loc[index,'4'])):
        dt = agentBreaks.loc[index,'3'].strftime("%d/%m/%Y")
        row=0
    
    if (not pd.isnull(agentBreaks.loc[index,'4']) and isinstance(agentBreaks.loc[index,'4'], str) and agentName != 'Undefined' and dt != 'Undefined'):
        if (agentBreaks.loc[index,'4'] !='Summary:'):
            #print(agentBreaks.loc[index,'4'])    
            tm=re.split(':', agentBreaks.loc[index,'5'])
            hr = int(tm[0])
            mn = int(tm[1])
            sc = int(tm[2])
            #print(agentName,dt,agentBreaks.loc[index,'3'],hr,mn,sc)
            mn=hr*60+sc/60+mn
            
            f.write("%s,%s,%s,%s\n"%(agentName,dt,agentBreaks.loc[index,'4'],mn)); f.flush()
            
    if (val=='Agent Name'):
        agentName=agentBreaks.loc[index,'3'][17:].split('-')[0]
        
    val=''
    row=row+1


f.close()
```
